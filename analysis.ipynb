```python
import json
from matplotlib import pyplot as plt
import os 
import sys 
from pathlib import Path
```

```python
run_logs = Path("./runs/")
weights_dict = {}
for fileName in os.listdir(run_logs):
    weight_path = run_logs.joinpath(fileName)
    with weight_path.open('r') as f:
        weights_dict[fileName.split('.')[0]] = json.load(f)
```

```python
for weight_name, values in weights_dict.items():
    x = []
    y = []
    for weight, info in values.items():
        x.append(float(weight.split(":")[1]))
        y.append(info[1][0])
    plt.plot(x, y)
    plt.xlabel("weight")
    if x[1] == x[0]*10:
        plt.xscale('symlog')
    plt.ylabel("wins in 500 games")
    plt.title(weight_name)
    plt.show()
```

```python
x = []
y = []
for weight, info in weights_dict['hand_synergy_up'].items():
    x.append(float(weight.split(":")[1]))
    y.append(info[1][0])
plt.plot(x, y)
plt.xlabel("weight")
plt.xscale('log')
plt.ylabel("wins in 500 games")
plt.title('hand_synergy_up')
plt.show()
```
